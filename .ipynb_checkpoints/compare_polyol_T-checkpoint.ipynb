### Henry's Constant

Henry's law states that the solubility of a gas in a liquid increases linearly with pressure at low pressures. This linear relationship is governed by the constant known as Henry's constant, given by the equation

\begin{align*}
    c &= H p\\
\end{align*}

where $c$ is the concentration of the gas in the liquid in $mol/m^3$, $p$ is the pressure in Pa, and $H$ is Henry's constant in $mol/(m^3\,Pa)$. To compute Henry's constant, I will first need to compute the concentration of $CO_2$ in the polyol in terms of moles per volume.

```python
reload(dataproc)
reload(plot)
# force fit through origin?
force_origin = True
# maximum pressure of data for fitting [kPa]
p_thresh = 1000
ax = None
color_list = ['r', 'b']
marker_list = ['x', 'o', '^']
tag_list = ['1k2f_30c', '1k3f_30c', '3k2f_31c', '1k2f_60c', '1k3f_60c', '3k2f_60c_lop']
# compute Henry's constant under each condition
for i in range(len(tag_list)):
    tag = tag_list[i]
    # increase threshold for pressure to 1000 kPa for fitting of Henry's constant
    H, s_H = dataproc.compute_henrys_const(d[tag]['p'], d[tag]['solub'], d[tag]['spec_vol'], p_thresh=p_thresh, 
                                      s_solub=d[tag]['s_solub'], s_spec_vol=d[tag]['s_spec_vol'], force_origin=force_origin)
    # convert to mmol
    H_mmol = H*1000
    s_H_mmol = s_H*1000
    T = d[tag]['T']
    polyol = d[tag]['polyol']
    # index for selecting marker
    ind_mk = int(polyol[0]) + int(polyol[2]) - 3
    # plot
    ax = plot.plot_errorbars(T, H_mmol, s_H_mmol, color=color_list[T < 60], marker=marker_list[ind_mk], 
                             xlabel='temperature [C]', ylabel='H [mmol/(m^3 Pa)]', label=tag,
                             title='Henry\'s Constant vs. Temperature', ax=ax)
    ax.set_ylim([0, 1])
    print('Henry''s constant for {0} is {1} mol/(m^3.Pa).'.format(tag, H))
    
plt.legend()

if save_plots:
    save_name = 'henrys_const_T'
    if force_origin:
        save_name += '_force_origin'
    else:
        save_name += '_free_yint'
    plt.savefig(data_folder + save_name + '.pdf', bbox_inches='tight')
```

This is interesting: even though the difference in solubility is not very clear among the polyols, the difference in Henry's constant shows a clear trend of increasing with the effective concentration of hydroxyl groups. Perhaps the effect of the hydroxyl groups is strongest at low CO2 concentrations because at higher CO2 concentrations they are "diluted" by the CO2?

In the plots above, the specific volume is compared for two difunctional, PPG-based polyols of different molecular weights to demonstrate the effect of molecular weight on the specific volume. The statistical uncertainty is shown below the data points and the systematic uncertainty is above. The reason for the dramatically larger systematic error is primarily the uncertainty in the density of the dry polyol. In both cases, the specific volume was greater for a longer molecular weight, although this was true both with and without CO2. We also see that the qualitative trends match for both molecular weights. At low temperature ($\approx 30$ C), the specific volume follows a non-monotonic trend as a function of the pressure of the surrounding CO2 atmosphere, initially decreasing and subsequently increasing. This behavior was seen previously in the Di Maio group in PCL samples under a CO2 atmosphere at similar pressures (Carbone et al., JSC, 2012). This non-monotonic trend is not observed for either polyol at the higher temperature ($60$ C), where the specific volume monotonically increases with CO2 pressure. The qualitative agreement for both polyols suggests that molecular weight has a quantitative--but not a qualitative--effect on the specific volume. In this case, the specific volume increases, but the systematic uncertainty in the dry polyol density makes this observation inconclusive.

Finally, we plot all the specific volume data to show the effect of temperature. Quantitatively, the specific volume of both the neat polyol and the polyol with CO2 increase with temperature, despite the lower solubility of CO2 in polyol. One might expect that at some pressure, the reduction of solubility of CO2 at higher temperatures might counter the reduction in the specific volume of the neat polyol, but for the pressures considered this is not true. Qualitatively, the effect of pressure on the specific volume varies dramatically between the low ($\approx 30$ C) and high ($60$ C) temperatures. At the low temperature, the specific volume varies non-monotonically with the pressure of the CO2 atmosphere, initially decreasing and subsequently increasing, as observed by Carbone et al. JSC, 2012 for PCL/CO2 mixtures. At the high temperature, the specific volume increases monotonically with the pressure for all three polyols, appearing to increase slightly more sharply about 6000 kPa. In Carbone et al. (2012), the non-monotonic trend for PCL/CO2 mixtures was observed more strongly at the higher temperatures (at least 90 C), contrary to our findings for polyol/CO2 mixtures. Although we cannot explain this discrepancy, the statistical uncertainty is low enough that we believe the difference in qualitative trends at different temperatures is significant, a result further strengthened by observing this difference consistently among three different polyols.

We first note that the diffusivity generally increases with the pressure of the CO2 atmosphere, which we expect since CO2 diffuses more easily in CO2 than in polyol. Nevertheless, the sparsity and large spread of our data at pressures above 5000 kPa does not indicate a clear trend. 

While we cannot distinguish a clear effect of the molecular weight or the functionality on the diffusivity above the uncertainty, we can see that the diffusivity is generally higher for higher temperatures.
