# Compare Polyol and Temperature

## Introduction

In this script, we compare the measured physical properties (solubility, specific volume, interfacial tension, and diffusivity) of CO2 in three different polyols (1k2f, 1k3f, and 3k2f, where the "k" signifies the molecular weight in kg/mol and the "f" signifies the functionality, that is, the number of -OH groups per molecule) at two different temperatures (30-31 C and 60 C).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import dataproc
import plot

from importlib import reload

# System parameters
# molecular weight of CO2
mw_co2 = 44.01
# conversion of m3 per mL
m3_per_mL = 1E-6
# Save plots?
save_plots = True
# file path to saved data
data_folder = '../g-adsa_results/'
# csv data files
csv_file_list = ['1k2f_30c', '1k3f_30c', '1k2f_60c', '1k3f_30c', '1k3f_60c', '3k2f_31c', '3k2f_60c_lop', '3k2f_60c_hip']
```

Extract data from dataframes

```python
d = {}
for i in range(len(csv_file_list)):
    file_name = csv_file_list[i]
    d[file_name] = {}
    d[file_name]['T'] = int(file_name[5:7])
    # add 0.5 c if the temperature is 30 (not included in file name)
    d[file_name]['T'] += 0.5*(d[file_name]['T'] == 30)
    d[file_name]['T'] += 0.1*(d[file_name]['T'] == 31)
    d[file_name]['polyol'] = file_name[:4]
    # get full data to store
    df = pd.read_csv(data_folder + file_name + '.csv')
    # store full data
    d[file_name]['p'] = df['p actual [kPa]'].to_numpy(dtype=float)
    d[file_name]['solub'] = df['solubility [w/w]'].to_numpy(dtype=float)
    d[file_name]['s_solub'] = df['solubility error [w/w]'].to_numpy(dtype=float)
    d[file_name]['w_gas'] = df['actual weight of dissolved gas [g]'].to_numpy(dtype=float)
    d[file_name]['v_samp'] = df['sample volume [mL]'].to_numpy(dtype=float)
    d[file_name]['spec_vol'] = df['specific volume [mL/g]'].to_numpy(dtype=float)
    d[file_name]['s_spec_vol'] = df['specific volume error [mL/g]'].to_numpy(dtype=float)
    d[file_name]['if_tension'] = df['if tension [mN/m]'].to_numpy(dtype=float)
    d[file_name]['s_if_tension'] = df['if tension std [mN/m]'].to_numpy(dtype=float)    
```

### Henry's Constant

Henry's law states that the solubility of a gas in a liquid increases linearly with pressure at low pressures. This linear relationship is governed by the constant known as Henry's constant, given by the equation

\begin{align*}
    c &= H p\\
\end{align*}

where $c$ is the concentration of the gas in the liquid in $mol/m^3$, $p$ is the pressure in Pa, and $H$ is Henry's constant in $mol/(m^3\,Pa)$. To compute Henry's constant, I will first need to compute the concentration of $CO_2$ in the polyol in terms of moles per volume.

```python
reload(dataproc)
ax = None
color_list = ['r', 'b']
marker_list = ['x', 'o', '^']
tag_list = ['1k2f_30c', '1k3f_30c', '3k2f_31c', '1k2f_60c', '1k3f_60c', '3k2f_60c_lop']
# compute Henry's constant under each condition
for i in range(len(tag_list)):
    tag = tag_list[i]
    H = dataproc.compute_henrys_const(d[tag]['p'], d[tag]['w_gas'], d[tag]['v_samp'])
    T = d[tag]['T']
    polyol = d[tag]['polyol']
    ind_mk = int(polyol[0]) + int(polyol[2]) - 3
    ax = plot.plot_line(d[tag]['T'], H, color=color_list[T < 60], marker=marker_list[ind_mk], xlabel='temperature [C]', 
                        ylabel='H [mol/(m^3 Pa)]', label=tag,
                       title='Henry''s Constant vs. Temperature', ax=ax)
    print('Henry''s constant for {0} is {1} mol/(m^3.Pa).'.format(tag, H))
    
plt.legend()

if save_plots:
    plt.savefig(data_folder + 'henrys_const_T.pdf', bbox_inches='tight')
```

This is interesting: even though the difference in solubility is not very clear among the polyols, the difference in Henry's constant shows a clear trend of increasing with the effective concentration of hydroxyl groups. Perhaps the effect of the hydroxyl groups is strongest at low CO2 concentrations because at higher CO2 concentrations they are "diluted" by the CO2?

#### Plot as function of density of CO2

```python
tag_list = ['1k2f_30c', '1k3f_30c', '3k2f_31c', '1k2f_60c', '1k3f_60c', '3k2f_60c_lop', '3k2f_60c_hip']
color_list = ['b', 'c', 'k', 'r', 'm', 'g', 'g']

for i in range(len(tag_list)):
    tag = tag_list[i]
    color = color_list[i]
    if i == 0:
        ax = plot.plot_errorbars_ads_des(dataproc.rho_co2(d[tag]['p'], d[tag]['T']),
                                         d[tag]['if_tension'], d[tag]['s_if_tension'], 
                                         d[tag]['p'], d[tag]['T'], 
                                         color=color, xlabel=r'$\rho_{CO2}$ [g/mL]', ylabel='Interfacial Tension [mN/m]', 
                                         title='Interfacial Tension vs. CO2 Density, Compare T', label_tag=tag[:4])
    else:
        ax = plot.plot_errorbars_ads_des(dataproc.rho_co2(d[tag]['p'], d[tag]['T']), d[tag]['if_tension'], 
                                         d[tag]['s_if_tension'], d[tag]['p'], d[tag]['T'], 
                                         color=color, ax=ax, label_tag=tag[:4])

# put legend outside of plot box
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width*0.65, box.height])
legend_x = 1
legend_y = 0.5
plt.legend(loc='center left', bbox_to_anchor=(legend_x, legend_y))# save figure

if save_plots:
    plt.savefig(data_folder + 'if_tension_effect_T_rho_co2.pdf', bbox_inches='tight')
```

The above plot shows that the difference in slope between the polyols at ~30 C and at 60 C is smaller when plotting the interfacial tension as a function of the density of the CO2 gas than when plotted as a function of the pressure. This decrease in the difference of slope indicates that the difference in slope is predominantly caused by the slower increase in CO2 density as a function of pressure at 60 C than near 30 C (see NIST data: https://webbook.nist.gov/cgi/fluid.cgi?T=333&PLow=0&PHigh=6&PInc=0.1&Applet=on&Digits=5&ID=C124389&Action=Load&Type=IsoTherm&TUnit=K&PUnit=MPa&DUnit=g%2Fml&HUnit=kJ%2Fmol&WUnit=m%2Fs&VisUnit=uPa*s&STUnit=N%2Fm&RefState=DEF and https://webbook.nist.gov/cgi/fluid.cgi?T=303&PLow=0&PHigh=6&PInc=0.1&Applet=on&Digits=5&ID=C124389&Action=Load&Type=IsoTherm&TUnit=K&PUnit=MPa&DUnit=g%2Fml&HUnit=kJ%2Fmol&WUnit=m%2Fs&VisUnit=uPa*s&STUnit=N%2Fm&RefState=DEF). Because the CO2 is not increasing in density as quickly with pressure at 60 C as near 30 C, the gaseous CO2 phase is not becoming as similar in density to the liquid phase as quickly with pressure, so the slope is lower at 60 C when plotted as a function of pressure. Nevertheless, we still see a slightly slower decrease in interfacial tension with the density of the CO2 phase at 60 C than near 30 C, indicating that the lower solubility of CO2 at 60 C than at 30 C is still in part responsible for the slower decrease in interfacial tension.
