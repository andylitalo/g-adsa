# Sensitivity

This notebook explores the sensitivity of theoretical predictions of different thermophysical properties, such as interfacial tension and solubility, to perturbations in the parameters of the models on which they are based. The models we explore are based on PC-SAFT and DFT and were largely developed by Dr. Huikuan Chao.

We see that increasing both $\epsilon$ and $\sigma$ by the right amount can cancel out their contrary effects on the solubility. Let's check how they affect the interfacial tension.

```python
# plots result of varying sigma
_ = plot.sensitivity_manual(d, d_dft, '3k2f_31c', 'if_tension', 'sigma', 3.01, data_folder, 
                            '3k2f_31c_if_tension_dft_vary_sigma_multi.pdf', 
                            ['sigma_2-86', 'sigma_3-07', 'sigma_3-16', 'epsn_240-0~sigma_3-07'])
```

**WOW!** The interfacial tension is also essentially unchanged even though the values of $\epsilon$ and $\sigma$ have both been increased. Perhaps this could resolve our problem with the specific volume, so I will plot the specific volume predictions in comparison to the measured values next. 

```python
reload(plot)

# plots result of varying sigma
_ = plot.sensitivity_manual(d, d_dft, '3k2f_31c', 'spec_vol', 'sigma', 3.01, data_folder, 
                            '3k2f_31c_if_tension_dft_vary_sigma_multi.pdf', 
                            ['sigma_2-86', 'sigma_3-07', 'sigma_3-16', 'epsn_240-0~sigma_3-07'])
```

**Great!** The concomitant increase in $\epsilon$ and $\sigma$ resulted in a more accurate prediction for the specific volume!

Let's see if we can get even closer. This increase was by 0.06 in $\sigma$ and by 11.5 in $\epsilon$, and it increased the specific volume by about 0.05 mL/g. To match the experimental measurements, we need to increase the specific volume by an additional 0.08 mL/g roughly. Assuming linear relationships, we might be able to achieve this by increasing $\sigma$ an additional 0.1 to have $\sigma = 3.17$ and $\epsilon$ by an additional 18 to have $\epsilon = 258~k_B$.

```python
# plots result of varying sigma
_ = plot.sensitivity_manual(d, d_dft, '3k2f_31c', 'solub', 'sigma', 3.01, data_folder, 
                            '3k2f_31c_solub_dft_vary_sigma_multi.pdf', 
                            ['sigma_2-86', 'sigma_3-07', 'sigma_3-16', 'epsn_240-0~sigma_3-07',
                            'epsn_258-0~sigma_3-17', 'epsn_240-0'])

# plots result of varying sigma
_ = plot.sensitivity_manual(d, d_dft, '3k2f_31c', 'if_tension', 'sigma', 3.01, data_folder, 
                            '3k2f_31c_if_tension_dft_vary_sigma_multi.pdf', 
                            ['sigma_2-86', 'sigma_3-07', 'sigma_3-16', 'epsn_240-0~sigma_3-07',
                            'epsn_258-0~sigma_3-17'])

# plots result of varying sigma
_ = plot.sensitivity_manual(d, d_dft, '3k2f_31c', 'spec_vol', 'sigma', 3.01, data_folder, 
                            '3k2f_31c_if_tension_dft_vary_sigma_multi.pdf', 
                            ['sigma_2-86', 'sigma_3-07', 'sigma_3-16', 'epsn_240-0~sigma_3-07',
                            'epsn_258-0~sigma_3-17'])
```

**Hypothesis confirmed!** Julie was onto something. By increasing $\sigma$ and $\epsilon$ further, I maintained the same DFT predictions for interfacial tension and solubility as for the fitted parameter values, but matched the specific volume much more accurately.

**The qualitative trend for the specific volume does not match the data, however.** The DFT predicts that the specific volume decreases with pressure, whereas measurements showed that specific volume generally increases with pressure. The error between the DFT prediction and the measurements is beyond the error bars, unfortunately.

The next test of this hypothesis is to see if it holds at the higher temperature (60 C).
