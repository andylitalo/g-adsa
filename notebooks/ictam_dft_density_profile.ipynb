# Plot Density Profile from DFT Calculations for ICTAM

```python
import numpy as np
import matplotlib.pyplot as plt

dft_pred_folder = '../../../Wang/dft_string_method/bi_surf/3k2f_31c_sensitivity/epsn_258-0~sigma_3-17/'
tk_fs = 18
ax_fs = 20
lw = 3
lw_dash = 2
ms = 7
c_co2 = '#FF3A3A'
c_3k2f = '#0047F2'
x_min = -25
x_max = 25

fig = plt.figure()
ax = fig.add_subplot(111)

# CO2 data
data_co2 = np.genfromtxt(dft_pred_folder + 'rho0_4.dat')
x = data_co2[:,0]
y = data_co2[:,1]
ax.plot(x - np.mean(x), y, label=r'CO$_2$', color=c_co2, lw=lw)

ax.plot([x_min, 0], [y[0], y[0]], color=c_co2, lw=lw_dash, linestyle='--')
ax.plot(0, y[0], color=c_co2, marker='o', ms=ms)
ax.plot([0, x_max], [y[-1], y[-1]], color=c_co2, lw=lw_dash, linestyle='--')
ax.plot(0, y[-1], color=c_co2, marker='o', ms=ms)

# polyol data
data_3k2f = np.genfromtxt(dft_pred_folder + 'rho1_4.dat')
x = data_3k2f[:,0]
y = data_3k2f[:,1]
ax.plot(x - np.mean(x), y, label='Polyol', color=c_3k2f, lw=lw)

# plots bulk values

ax.plot([x_min, 0], [y[0], y[0]], color=c_3k2f, lw=lw_dash, linestyle='--')
ax.plot(0, y[0], color=c_3k2f, marker='o', ms=ms)
ax.plot([0, x_max], [y[-1], y[-1]], color=c_3k2f, lw=lw_dash, linestyle='--')
ax.plot(0, y[-1], color=c_3k2f, marker='o', ms=ms)
    
ax.set_xlabel(r'distance [$\AA$]', fontsize=ax_fs)
ax.set_ylabel(r'conc. [beads/$\sigma^3$]', fontsize=ax_fs)
ax.tick_params(which='both', labelsize=tk_fs)
ax.set_xlim([x_min, x_max])
ax.set_ylim([0, 0.6])
```
