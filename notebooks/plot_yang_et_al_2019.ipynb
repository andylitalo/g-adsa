# Compare Polyol and Temperature for Data from Yang et al. (2019)

## Introduction

In this notebook, we compare the measured physical properties (solubility, specific volume, interfacial tension, and diffusivity) of CO2 in seven different polyols (800-5f, 500-4f, 1k3f, 450-3f, 255-3f, 1k2f, 356-2f, where the "k" signifies the molecular weight in kg/mol and the "f" signifies the functionality, that is, the number of -OH groups per molecule) at 35 C, although 800-5f and 500-4f also have data at 50 C.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import dataproc
import plot

from importlib import reload

reload(dataproc)

# System parameters
# molecular weight of CO2
mw_co2 = 44.01
# conversion of m3 per mL
m3_per_mL = 1E-6
# Save plots?
save_plots = True
# file path to saved data
data_folder = '../g-adsa_results/'
# csv data files
csv_file_list = ['p1_800-5f_35c', 'p1_800-5f_50c', 'p2_500-4f_35c', 'p2_500-4f_50c', 'p3_1k3f_35c', 'p4_450-3f_35c',
                 'p5_255-3f_35c', 'p6_1k2f_35c', 'p7_356-2f_35c']
# folder of csv data files for DFT predictions of interfacial tension
dft_data_folder = 'dft_pred//'
# folder of csv data files showing sensitivity of DFT predictions to PC-SAFT parameters
dft_sensitivity_folder = 'dft_pred//3k2f_31c_sensitivity//'
```

Extract data from dataframes

```python
d = dataproc.load_proc_data(csv_file_list, data_folder)
```

### Compare Effects of Experimental Variables

#### Effect of Molecular Weight

```python
# molecular weight (35 C, 3f)
tag = 'p3_1k3f_35c'
ax = plot.plot_errorbars_ads_des(d[tag]['p'], d[tag]['solub'], d[tag]['s_solub'], 
                                d[tag]['p'], d[tag]['T'], color='r', xlabel='p [kPa]',
                                ylabel='solubility [w/w]', title='Solubility vs. p, Vary Mw (3f, 35 C)',
                                label_tag='1000 g/mol')
tag = 'p4_450-3f_35c'
ax = plot.plot_errorbars_ads_des(d[tag]['p'], d[tag]['solub'], d[tag]['s_solub'], 
                                d[tag]['p'], d[tag]['T'], color='g', ax=ax,
                                 label_tag='450 g/mol')
tag = 'p5_255-3f_35c'
ax = plot.plot_errorbars_ads_des(d[tag]['p'], d[tag]['solub'], d[tag]['s_solub'], 
                                d[tag]['p'], d[tag]['T'], color='b', ax=ax,
                                 label_tag='255 g/mol')
ax.set_ylim([0, ax.get_ylim()[1]])
plt.legend()
# save figure
if save_plots:
    plt.savefig(data_folder + 'solub_yang_mw_3f_35c.pdf', bbox_inches='tight')
    
# molecular weight (35 C, 2f)
tag = 'p6_1k2f_35c'
ax = plot.plot_errorbars_ads_des(d[tag]['p'], d[tag]['solub'], d[tag]['s_solub'], 
                                d[tag]['p'], d[tag]['T'], color='r', xlabel='p [kPa]',
                                ylabel='solubility [w/w]', title='Solubility vs. p, Vary Mw (2f, 35 C)',
                                label_tag='1000 g/mol')
tag = 'p7_356-2f_35c'
ax = plot.plot_errorbars_ads_des(d[tag]['p'], d[tag]['solub'], d[tag]['s_solub'], 
                                d[tag]['p'], d[tag]['T'], color='b', ax=ax,
                                 label_tag='356 g/mol')
ax.set_ylim([0, ax.get_ylim()[1]])
plt.legend()
# save figure
if save_plots:
    plt.savefig(data_folder + 'solub_yang_mw_2f_35c.pdf', bbox_inches='tight')
```

For both 2-functional and 3-functional polymers, Yang et al. (2019) found that increasing the molecular weight significantly increases the solubility in the pressure range 2-6 MPa at 35 C. This is opposite the finding from the DFT model fitted to the G-ADSA data, which found that increasing the $N$ parameter, which is proportional to the molecular weight, decreases the solubility.

I also want to point out that the error bars are unrealistically small, so as to be invisible for some of the data points.

Where the molecular weight is exactly the same (1kg/mol), increasing the functionality from 2f to 3f slightly increased the solubility, in contrast to our findings with G-ADSA in Naples.

If the *functionality decreases* and the *molecular weight increases* (5.4 -> 2 or 3, 800 g/mol -> 1000 g/mol), the solubility significantly decreases (top plot). If the *functionality and molecular weight increase* (3f->4f, 450 g/mol -> 500 g/mol), the solubility also significantly decreases (bottom plot). However, we saw earlier that when just the molecular weight increases, the solubility increases. There seems to be at least one contradiction among these results.

Yang et al. (2019) found that the effect of increasing temperature 15 C from 35 C to 50 C on solubility was less than the effect of increasing molecular weight from 500->800 g/mol (which increases solublity) and the effect of increasing functionality from 4f->5.4f (unknown effect on solubility based on Yang et al.) combined.
