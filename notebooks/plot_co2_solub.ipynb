## Plot Solubility of CO2 in Glycerol

In our Zoom meeting on July 8, 2020, Julie suggested examining the effect of hydroxyl (-OH) concentration to its extremes, the upper end including glycerol, which is 3 hydroxyl groups connected on 3-carbon backbone, and ethylene glycol, which is 2 hydroxyl groups connected on a 2-carbon backbone. Data for solubility of CO2 in ethylene glycol are available in the literature, so I plot those data here for comparison to CO2 solubility in polyols with lower hydroxyl groups. This will test our hypothesis that a higher hydroxyl count increases CO2 solubility, which was based on Prof. Di Maio's finding with Raman spectroscopy that CO2 and hydroxyl groups interact strongly when CO2 is dissolved in polyol.
